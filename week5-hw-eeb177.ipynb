```python
def makeBirdDict(filename):
    # this function opens the data file from nature-birds and extracts the species and families as a dictionary
    BirdDict= {}
    fobj= open(filename, "r",encoding = "ISO-8859.15") # open the file
    all_records = fobj.readlines()[1:] # read the file (list)
    for line in all_records:
        record_elements = line.split(",") # split lines on comma (this produces another list)
        species = record_elements[0]
        family = record_elements[2]
        BirdDict[species] = family # stick species and family into birdDict

    return BirdDict
```

#### Percentage of amino acid residues, part one


```python
def get_aa_percentage(protein, aa):
    # convert both inputs to upper case
    protein = protein.upper()
    aa = aa.upper()
    aa_count = protein.count(aa)
    protein_length = len(protein)
    percentage = aa_count * 100 / protein_length
    return percentage

assert get_aa_percentage("MSRSLLLRFLLFLLLLPPLP", "M")
```

#### Percentage of amino acid residues, part two

```python
def get_aa_percentage(protein, aa_list=['A','I','L','M','F','W','Y','V']):
    protein = protein.upper()
    protein_length = len(protein)
    total = 0
    for aa in aa_list:
        aa = aa.upper()
        aa_count = protein.count(aa)
        total = total + aa_count
    percentage = total * 100 / protein_length
    return percentage

assert get_aa_percentage("MSRSLLLRFLLFLLLLPPLP", ["M"])
```

```python
get_aa_percentage("MSRSLLLRFLLFLLLLPPLP", ['F', 'S', 'L']) 
```
